```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import joblib
```

```python
# read file
path= 'mall_customer_seg/mall_cust_segmentation.csv'
df= pd.read_csv(path)
print(df.head())
print(df.tail())
```

```python
# sanity check (null, duplicates, garbage)
print(df.shape)
print(df.isnull().sum()/ df.shape[0]*100)
print(df.duplicated().sum())
for i in df.select_dtypes(include="object").columns:
 print(df[i].value_counts())
```

```python
# EDA
df.describe().T
# [Hisogram (distributuion of data)]
for i in df.select_dtypes('number').columns:
 sns.histplot(data=df, x=i)
 plt.show()
```

```python
# Boxplot (check outliers)
for i in df.select_dtypes('number').columns:
 sns.boxplot(data=df, x=i)
 plt.show()
```

```python
features= ['Age', 'Annual Income', 'Spending Score']
X= df[features]
```

```python
print("Features selected for clustering:")
print(X.head())
```

```python
#scaling features 
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
# finding optimal number of clusters k 
wcss = []
for i in range(1, 11): # Try K from 1 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
#elbow method graph 
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# choosing 4 as the optimal number of clusters
n = 4
kmeans = KMeans(n_clusters=n, init='k-means++', random_state=42, n_init='auto')

# Fitting the model to get the cluster assignments for each customer
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
print(f"Data with Cluster Assignments (K={n}):")
print(df.head())
```

```python
print(f"Average characteristics of each cluster (K={n}):")
cluster_summary = df.groupby('Cluster')[features].mean()
print(cluster_summary)
```

```python
 # Annual Income vs. Spending Score.
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Annual Income', y='Spending Score', hue='Cluster', data=df,
                palette='viridis', s=100, alpha=0.8, legend='full')
plt.title(f'Customer Segments (K={n}) by Annual Income vs. Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

```python
model_filename= 'mall_cust_seg.joblib'
```

```python
joblib.dump(kmeans, model_filename)
```
